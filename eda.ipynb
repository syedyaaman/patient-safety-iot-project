# Exploratory Data Analysis (EDA)
## Global Patient Safety Observatory Dataset

This notebook explores the Global Patient Safety Observatory dataset to understand
its structure, quality, and trends before modeling.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df_selected = df[
    ["Location", "Period", "Indicator", "Value"]
]

df_selected.head()
```

```python
df_selected.isnull().sum()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(df_selected["Period"], df_selected["Value"], marker="o")
plt.xlabel("Year")
plt.ylabel("Patient Safety Score")
plt.title("Patient Safety Score Over Time")
plt.show()
```

```python
df_selected.to_csv("patient_safety_clean.csv", index=False)
```
